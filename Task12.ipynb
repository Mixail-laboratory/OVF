```python
from math import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(t, a_0, a_1, w_0, w_1):
    return a_0 * sin(w_0*t) + a_1 * sin(w_1*t)
```

```python
def fourie_pow_spect(f, h, t_s, a0, a1, w0, w1):
    pow_spect = []
    w = []
    for i in range(n):
        fi = complex(0, 0)
        
        for k in range(n):
            fi += (1/n)*f(t_s[k], a0, a1, w0, w1)*np.exp(2*np.pi*(1j)*i*k/n)*h(k)
            
        pow_spect.append((fi*fi.conjugate()).real) #мощность - квадрат модуля
        w.append(2*np.pi*i/T)
        
    return w, pow_spect

def h_window(ks):
    return 1/2 * (1 - np.cos(2*pi*ks/n))

def rect_window(x):
    return 1

```

```python
a0 = 1
w0 = 5.1
a1 = 0.001
w1 = 25.5/2
n = 30
T = 2 * pi
t_step = T / n
W = 2 * pi / t_step
w_step = W / n
```

```python
time_array = np.linspace(0, T, n)

plt.figure(figsize=(15, 6))
plt.plot(time_array, [func(t, a0, a1, w0, w1) for t in time_array], 'mo-')
plt.grid()
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('signal')
plt.show()
```

```python
w_no, spec_no = fourie_pow_spect(func, rect_window, time_array, a0, a1, w0, w1)

plt.figure(figsize=(15, 6))
plt.plot(w_no, spec_no, 'mo-')
plt.grid()
plt.yscale('log')
plt.xlabel('w')
plt.ylabel('|f(w)|^2')
plt.title('With rectangle window')
plt.show()
```
